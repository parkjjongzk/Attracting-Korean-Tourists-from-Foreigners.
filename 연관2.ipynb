```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
df = pd.read_excel("C:/Users/pc/Desktop/연관.xlsx")
```

```python
df=df.astype(object)
```

```python
list=df.values.tolist()
```

```python
list
len(list)
```

```python
for idxx, i in enumerate(list):
    if idxx==16037:
        break
    else:
        a=len(i)
        for idx, j in enumerate(i):
            if idx <= a:
                if j==5:
                    i.remove(5)  #빈값을 5로 바꾸고 제거
                else:
                    continue
            else:
                break
```

```python
te = TransactionEncoder()
te = te.fit(list)
```

```python
te = TransactionEncoder()
te_ary = te.fit(list).transform(list)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df, min_support=0.005, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.005) 
```
